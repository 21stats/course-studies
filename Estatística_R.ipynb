```r
library(readxl)
if (!require (data.table)){ install.packages('data.table')}
library(dplyr)
library(data.table)
```

```r
acidentes_SP = read.csv('/content/datatran_SP.csv')
acidentes_SP
```

```r
km = as.double(acidentes_SP$km)
km_filtrado = na.omit(km)
km_filtrado
```

###Frequências:

```r
freq_mortos = table(acidentes_SP$mortos)
```

```r
freq_mortos
```

```r
freq_mortos_ta = prop.table(table(acidentes_SP$mortos))
freq_mortos_ta
```

```r
sum(acidentes_SP$mortos)
```

```r
freq_rel_mortos = 100 * prop.table(table(acidentes_SP$mortos))
freq_rel_mortos = c(round(c(freq_rel_mortos, sum(freq_rel_mortos)), digits = 2))
names(freq_rel_mortos)[7] = 'Total'
```

```r
freq_rel_mortos
```

```r
#adicionar uma coluna

freq_mortos_t = c(freq_mortos, sum(acidentes_SP$mortos))
names(freq_mortos_t)[7] = 'Total' #alterar o nome da coluna

freq_mortos_t = as.data.frame(freq_mortos_t)
freq_mortos_t
```

```r
freq_km = table(acidentes_SP$km)
freq_km
```

```r
intervalo_km = seq(from = 0.0,to = 600.0, by =100.0)

tabela_estradaBr = table(cut(km_filtrado, breaks = intervalo_km, right = FALSE))
tabela_estradaBr
```

```r
freq_veiculos = table(acidentes_SP$veiculos)
freq_veiculos = as.vector(freq_veiculos)
freq_veiculos
```

### Medidas de tendência central e de dispersão

```r
km_filtrado
```

```r
print(paste0('Média: ',mean(km_filtrado)))
```

```r
print(paste0('Mediana: ',median(km_filtrado)))
```

```r
moda = function(v) {
  valor_unico = unique(v)
  valor_unico[which.max(tabulate(match(v, valor_unico)))]
}
```

```r
print(paste0('Moda: ',moda(km_filtrado)))
```

```r
print(paste0('Desvio padrão: ', round(sd(km_filtrado), digits = 2)))
```

```r
print(paste0('Variância: ', var(km_filtrado)))
```
