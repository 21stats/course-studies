```python
FAAMG = ['META', 'AMZN', 'AAPL', 'MSFT', 'GOOG']
```

```python
ações_df = pd.DataFrame()
for ação in FAAMG:
  ações_df[ação] = yf.download(ação, start = '2023-01-01')['Close']
```

```python
ações_df
```

```python
ações_df.isnull().sum()
```

```python
np.arange(0, len(ações_df.columns))
```

```python
plt.figure(figsize = (5,25))
i = 1 # Começa na coluna 1, a coluna 0 é a data, que iremos ignorar.
for i in np.arange(0, len(ações_df.columns)):
  plt.subplot(6, 1, i + 1)
  sns.histplot(ações_df[ações_df.columns[i]])
```

```python
ações_df.plot(figsize = (10,5), title = 'FAAMG prices at time')
```

```python
ações_df_normalizado = ações_df.copy()
```

```python
i = 1
for i in ações_df_normalizado.columns[0:]:
  ações_df_normalizado[i] = ações_df_normalizado[i] / ações_df_normalizado[i][0]
```

```python
ações_df_normalizado
```

```python
ações_df_normalizado.isnull().sum()
```

```python
ações_df_normalizado.plot(figsize = (10,5), title = 'FAAMG prices at time - normalizado')
```

```python
dinamic_graph = px.line(ações_df_normalizado)
for i in ações_df.columns[1:]:
   dinamic_graph.add_scatter(x = ['2023-01-01', '2023-09-02'], y = ações_df_normalizado[i])
```

```python
dinamic_graph.show()
```
