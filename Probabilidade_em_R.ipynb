```r
SetIntersection = function(A, B) {
  AinB = c()
  emptyset = '∅'

  for (i in 1:length(A)){
    for (j in 1:length(B)) {
      if (A[i] == B[j]) {
        AinB = append(AinB, A[i])
      }
    }
  }
    if (length(AinB) == 0) {
      return(emptyset)
  } else {
      return(AinB)
  }
}



SetUnion = function (A, B) {
  AuB = c(A, B) #concatenate
  AuB = unique(sort(AuB))
    return(AuB)
}



SetComplement = function (A, B, U = c(),
                          containsets = TRUE,
                          ans = '') {
  if (containsets== TRUE) {
    U = append(U, A)
    U = append(U, B)
  }
  AuB = unique(sort(c(A,B)))
  ordU = unique(sort(U))


  verify = TRUE

  n = 1
  for (v in 1:length(AuB)) {
    while (n <= length(ordU)) {
      if (AuB[v] != ordU[n]) {
        n = n + 1
      } else {
          break
          n = 1
      }
    }
    if (n > length(ordU)) {
      verify = FALSE
      break
    }
  }
  return(verify)

  # if (U < AuB) {
  #   print('Universal set is smaller than the union of sets A and B')
  # } else {
  #    result = switch(set,
  #    compA = c()
  #     'A' = {
  #     for (i in length(A)) {

  #    }
  #  })
  # }
  # result = switch(set,
  #   'A' = {
  #     for (i in length(A)) {

  #     }
  #   })
}
```
