```r
prob = function(A, B = 0, E,
                neg = FALSE,
                intersec = FALSE) {

  if (neg == FALSE && intersec == FALSE) {
    res = (A / E ) * 100
  }
  else if (neg == TRUE && intersec == FALSE) {
    res = 100 - ((A / E) * 100)
  }
  else if (neg == FALSE && intersec == TRUE) {
    res = ((A/E) * (B/E)) * 100
  }
  else if (neg == TRUE && intersec == TRUE) {
    res = 100 - ((A/E) * (B/E)) * 100
  }
  return(res)
}
```

```r
prob(A = nrow(class_feridos),
     E = nrow(acidentes_SP),
     neg = TRUE)
```
