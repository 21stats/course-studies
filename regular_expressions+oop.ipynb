```python
'''
Para revisar o conteúdo prático sobre expressões regulares, implemente o exercício abaixo.
Logo em seguida você pode acessar a aula em vídeo com a solução
Crie expressões regulares para extrair as seguintes informações do texto abaixo (use a função findall):
- Números
- CEPs
- URLs
'''
```

```python
informações = 'Matheus, (11)23598-1113, 12345678, www.miro.com'
```

```python
import re
```

```python
telefone = re.findall('\(\d{2}\)\d{5}-\d{4}', informações)
print(telefone)
```

```python
CEP = re.findall('\d{8}', informações)
print(CEP)
```

```python
URL = re.findall('\w+\.\w+\.\w+', informações)
print(URL)
```

```python
'''Crie uma classe chamada aluno com os seguintes atributos:
- Nome
- Nota 1
- Nota 2
- Crie um construtor para a classe (__init__)
Crie as seguintes funções (métodos):
- Calcula média, retornando a média aritmética entre as notas
- Mostra dados, que somente imprime o valor de todos os atributos
- Resultado, que verifica se o aluno está aprovado ou reprovado (se a média for maior ou igual a 6.0, o aluno está aprovado)
Crie dois objetos (aluno1 e aluno2) e teste as funções'''
```

```python
class aluno:
  def __init__(self,nome,nota1,nota2):
    self.nome = str(nome)
    self.nota1 = float(nota1)
    self.nota2 = float(nota2)

  def media(self):
    return (self.nota1 + self.nota2) / 2

  def dados(self):
    print('Nome: ', self.nome)
    print('Nota1: ', self.nota1)
    print('Nota2: ', self.nota2)

  def resultado(self):
    if self.media() >= 6:
      return 'aprovado'
    else:
      return 'reprovado'
```

```python
a1 = aluno('Agner', 10, 6)
```

```python
a1.media()
```

```python
a1.dados()
```

```python
a1.resultado()
```

```python
a2 = aluno('Bia',9,7)
```

```python
a2.media()
```

```python
a2.dados()
```

```python
a2.resultado()
```
